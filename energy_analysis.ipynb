```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

Getting data from imported csv file

```python
data = []
tab_len = 13
year_rows = [9, 25, 41]
for cnt, ele in enumerate(year_rows):
  year_data = pd.read_csv('/content/tarbimine.csv', skiprows=ele, nrows=tab_len, sep=';', decimal=',')
  year_data.columns.values[0] = f'202{cnt}'
  data.append(year_data)
  display(year_data)
```

Draw charts

```python
plt.figure(figsize=[15, 6.5])
plt.plot(data[0]['2020'].head(12), data[0]['Kokku (kWh)'].head(12), label='2020')
plt.plot(data[1]['2021'].head(12), data[1]['Kokku (kWh)'].head(12), label='2021')
plt.xticks(rotation=90)
plt.ylabel('Tarbimine [kWh]')
plt.xlabel('Kuu')
plt.legend()
plt.grid(b=True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
