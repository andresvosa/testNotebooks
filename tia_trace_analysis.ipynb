```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

Getting data from imported csv file

```python
trace_data = pd.read_csv('/content/TraceSpeedError_03.csv', sep=',', decimal='.')
trace_data
```

Draw charts

```python
plt.figure(figsize=[25, 6.5])
#plt.plot(trace_data['Sample'], trace_data['L21 Linespeed IN[%IW256]'], label='Input Raw speed')
plt.plot(trace_data['Sample'], trace_data['LineDB.Line general.Actual values.L21 actual speed'], label='Speed [m/s]')
plt.xticks(rotation=90)
plt.ylabel('Speed Recordings')
plt.xlabel('Sample No')
plt.legend()
plt.grid(b=True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
